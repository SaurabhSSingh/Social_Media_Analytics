```python
import json
from pprint import pprint
import nltk
stopwords1 = nltk.corpus.stopwords.words('english')
stopwords2 = ['rt', 'trump', 'trump\'s', '&amp;', '@trump', 'via', 'donald', 'it\'s', 'https', 'https:..', 'https:co', '@potus', 'https://t.co',
             '@realdonaldtrump', '@donaldtrump', '#trump', '\U0001f30e', 'he\'s']
stopwords = stopwords1 + stopwords2
words = []
delim = '\n'
with open('final_tweets_trump.json', 'r') as f:
    tweets = json.load(f)
    for s in tweets:
        s1 = s.strip()
        s2 = s1.lower()
        brk_word = s2.split(delim)
        for s3 in brk_word:
            word1 = s3.split(' ')
        words.append(word1)
print(len(words))
print(words[0])
for lst1000 in words:
    for remove_url in lst1000:
        if 'https:' in remove_url:
            lst1000.remove(remove_url)
    for remove_tco in lst1000:
        if 't.co' in remove_tco:
            lst1000.remove(remove_tco)
for lst1000 in words:
    for remove_url in lst1000:
        if 'https:' in remove_url:
            print(remove_url)
words2 = []
for lst1000 in words:
    store_list = []
    for w in lst1000:
        if w not in stopwords and len(w) > 1:
            store_list.append(w)
    words2.append(store_list)
print(len(words2))
print(words2[0])
```

```python
from nltk.stem.lancaster import LancasterStemmer
ls = LancasterStemmer()
words_clean = []
for ww in words2:
    words_clean.append(ls.stem(ww))
print len(words_clean)
```

```python
delimiter = ' '
words3 = []
for lst1000 in words2:
    words3 = words3+lst1000
    complete_string = delimiter.join(words3)
print (type(complete_string) , len(words3))
```

```python
from gensim import corpora
docs = words2
dic = corpora.Dictionary(docs)
print(dic)
```

```python
corpus = [dic.doc2bow(text) for text in docs]
print(type(corpus), len(corpus))
```

```python
for corp in corpus:
    print(len(corp), corp[:10])
```

```python
from gensim import models
tfidf = models.TfidfModel(corpus)
print(type(tfidf))
```

```python
corpus_tfidf = tfidf[corpus]
print(type(corpus_tfidf))
```

```python
NUM_TOPICS = 10
model = models.ldamodel.LdaModel(corpus_tfidf, 
                                 num_topics=NUM_TOPICS, 
                                 id2word=dic, 
                                 update_every=1, 
                                 passes=12)
```

```python
print("LDA model")
topics_found = model.print_topics(20)
counter = 1
for t in topics_found:
    print("Topic #{} {}".format(counter, t))
    counter += 1
```
